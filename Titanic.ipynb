*Add sibsp

```python
data = data[["sex","pclass","fare","age","sibsp","survived"]]
```

```python
data
```

```python
data = data.dropna()
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = data[["sex","pclass","fare","age","sibsp"]]
x_train,x_test,y_train,y_test = train_test_split(x,data.survived,test_size=0.2, random_state=0)
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train_norm = sc.fit_transform(x_train)
x_test_norm = sc.fit_transform(x_test)
```

```python
from sklearn import tree

DT = tree.DecisionTreeClassifier(criterion='entropy',max_depth=3)
DT
```

```python
DT.fit(x_train_norm,y_train)
```

```python
result = DT.predict(x_test_norm)
result
```

```python
from sklearn import metrics

metrics.accuracy_score(y_test,result)
```

```python
from sklearn.metrics import confusion_matrix

conf = confusion_matrix(y_test,result)
conf
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,result))
```
